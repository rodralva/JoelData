```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

$N_{fired}=N_{total}\times\left[1-\exp(-N_{seed}/N_{total}) \right]$

```python
N_total=2668
N_seed =np.arange(1,N_total)
N_fired=N_total*(1-np.exp(-N_seed/N_total))

fig, (ax1, ax2) = plt.subplots(2,1,figsize=(4, 6), dpi=100,gridspec_kw={'height_ratios': [3, 1]},sharex=True)

ax1.plot(N_seed,N_seed)
ax1.plot(N_seed,N_fired)
ax1.grid()
ax1.legend(["$N_{fired}=N_{seed}$","True fired cells"])
ax1.set_ylabel("$N_{fired}$")

ax2.plot(N_seed,N_seed/N_seed-1)
ax2.plot(N_seed,N_fired/N_seed-1)
# ax2.set_ylim([-.5,.5])
ax2.yaxis.set_major_locator(plt.MultipleLocator(0.1))
ax2.set_ylabel(r"$\frac{N_{fired}-N_{seed}}{N_{seed}}$")
ax2.set_xlabel(r"$N_{seed}$")
ax2.grid()

```

```python
%matplotlib inline

plt.figure(figsize=(8, 6), dpi=200);

for k in [100,200,1000]:
    ls=[]
    for j in range(10000):
        x=[int(random.uniform(1, N_total)) for i in range(k)];
        x=np.array(x)
        ls.append(len(np.unique(x)))
    plt.hist(ls,1000,range=[0,1000])
    plt.plot(N_fired[k-1],500,marker=".")
plt.semilogy()
```
